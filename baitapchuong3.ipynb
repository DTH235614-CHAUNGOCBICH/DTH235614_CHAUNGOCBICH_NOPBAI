```python
#cau3
import math

def giai_phuong_trinh_bac_2(a, b, c):
    """
    Giải phương trình bậc 2: ax² + bx + c = 0
    """
    if a == 0:
        # Trở thành phương trình bậc 1: bx + c = 0
        if b == 0:
            if c == 0:
                return "Phương trình có vô số nghiệm"
            else:
                return "Phương trình vô nghiệm"
        else:
            x = -c / b
            return f"Phương trình có 1 nghiệm: x = {x:.2f}"

    # Tính delta
    delta = b**2 - 4*a*c

    if delta < 0:
        return "Phương trình vô nghiệm"
    elif delta == 0:
        x = -b / (2*a)
        return f"Phương trình có nghiệm kép: x = {x:.2f}"
    else:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        return f"Phương trình có 2 nghiệm phân biệt:\nx1 = {x1:.2f}\nx2 = {x2:.2f}"

# Nhập hệ số từ bàn phím
try:
    a = float(input("Nhập hệ số a: "))
    b = float(input("Nhập hệ số b: "))
    c = float(input("Nhập hệ số c: "))

    ket_qua = giai_phuong_trinh_bac_2(a, b, c)
    print(f"\nKết quả: {ket_qua}")

except ValueError:
    print("Lỗi: Vui lòng nhập số hợp lệ!")
```

```python
#cau6
def doc_so_co_2_chu_so(n):
    """
    Đọc số có tối đa 2 chữ số thành chữ
    """
    # Danh sách cách đọc các số từ 0-9
    don_vi = ["không", "một", "hai", "ba", "bốn", "năm", "sáu", "bảy", "tám", "chín"]

    # Danh sách cách đọc hàng chục
    chuc = ["", "mười", "hai mươi", "ba mươi", "bốn mươi", "năm mươi",
            "sáu mươi", "bảy mươi", "tám mươi", "chín mươi"]

    # Kiểm tra số hợp lệ
    if n < 0 or n > 99:
        return "Số không hợp lệ! Vui lòng nhập số từ 0 đến 99"

    # Xử lý số có 1 chữ số (0-9)
    if n < 10:
        return don_vi[n]

    # Xử lý số có 2 chữ số
    hang_chuc = n // 10
    hang_don_vi = n % 10

    # Các trường hợp đặc biệt
    if hang_chuc == 1:  # Số từ 10-19
        if hang_don_vi == 0:
            return "mười"
        elif hang_don_vi == 5:
            return "mười lăm"
        else:
            return f"mười {don_vi[hang_don_vi]}"

    elif hang_don_vi == 0:  # Số tròn chục (20, 30, ..., 90)
        return chuc[hang_chuc]

    elif hang_don_vi == 1:  # Số có đơn vị là 1 (21, 31, ..., 91)
        return f"{chuc[hang_chuc]} mốt"

    elif hang_don_vi == 5:  # Số có đơn vị là 5 (25, 35, ..., 95)
        return f"{chuc[hang_chuc]} lăm"

    else:  # Các trường hợp còn lại
        return f"{chuc[hang_chuc]} {don_vi[hang_don_vi]}"

def main():
    """
    Hàm chính của chương trình
    """
    print("CHƯƠNG TRÌNH ĐỌC SỐ CÓ TỐI ĐA 2 CHỮ SỐ")
    print("=" * 40)

    try:
        # Nhập số từ bàn phím
        n = int(input("Nhập một số (0-99): "))

        # Đọc số và in kết quả
        ket_qua = doc_so_co_2_chu_so(n)
        print(f"Số {n} đọc là: {ket_qua}")

    except ValueError:
        print("Lỗi: Vui lòng nhập một số nguyên hợp lệ!")

# Chạy chương trình
if __name__ == "__main__":
    main()
```

```python
#cau7
def la_nam_nhuan(nam):
    """Kiểm tra xem một năm có phải là năm nhuận không"""
    if (nam % 4 == 0 and nam % 100 != 0) or (nam % 400 == 0):
        return True
    return False

def so_ngay_trong_thang(thang, nam):
    """Trả về số ngày trong tháng của một năm"""
    if thang in [1, 3, 5, 7, 8, 10, 12]:
        return 31
    elif thang in [4, 6, 9, 11]:
        return 30
    elif thang == 2:
        if la_nam_nhuan(nam):
            return 29
        else:
            return 28
    return 0

def kiem_tra_ngay_hop_le(ngay, thang, nam):
    """Kiểm tra xem ngày tháng năm có hợp lệ không"""
    if nam < 1:
        return False
    if thang < 1 or thang > 12:
        return False
    so_ngay_max = so_ngay_trong_thang(thang, nam)
    if ngay < 1 or ngay > so_ngay_max:
        return False
    return True

def tim_ngay_ke_tiep(ngay, thang, nam):
    """Tìm ngày kế tiếp của một ngày cho trước"""
    if not kiem_tra_ngay_hop_le(ngay, thang, nam):
        return "Ngày không hợp lệ!"

    so_ngay_trong_thang_hien_tai = so_ngay_trong_thang(thang, nam)

    # Trường hợp thông thường: chưa phải cuối tháng
    if ngay < so_ngay_trong_thang_hien_tai:
        return ngay + 1, thang, nam

    # Trường hợp cuối tháng nhưng chưa phải tháng 12
    elif thang < 12:
        return 1, thang + 1, nam

    # Trường hợp cuối năm: 31/12
    else:
        return 1, 1, nam + 1

def main():
    # Nhập ngày từ người dùng
    try:
        ngay = int(input("Nhập ngày: "))
        thang = int(input("Nhập tháng: "))
        nam = int(input("Nhập năm: "))

        # Kiểm tra tính hợp lệ của ngày nhập
        if not kiem_tra_ngay_hop_le(ngay, thang, nam):
            print("Ngày không hợp lệ!")
            return

        # Tìm ngày kế tiếp
        ngay_ke, thang_ke, nam_ke = tim_ngay_ke_tiep(ngay, thang, nam)

        # Hiển thị kết quả
        print(f"Ngày hiện tại: {ngay:02d}/{thang:02d}/{nam}")
        print(f"Ngày kế tiếp: {ngay_ke:02d}/{thang_ke:02d}/{nam_ke}")

    except ValueError:
        print("Vui lòng nhập số nguyên!")

# Chạy chương trình
if __name__ == "__main__":
    main()
```

```python
#cau8
# Nhập hai số a và b từ bàn phím
a = float(input("Nhập số a: "))
b = float(input("Nhập số b: "))

# Nhập ký tự ch từ bàn phím
ch = input("Nhập ký tự toán tử (+, -, *, /): ")

# Kiểm tra và thực hiện phép tính tương ứng
if ch == '+':
    ket_qua = a + b
    print(f"{a} + {b} = {ket_qua}")
elif ch == '-':
    ket_qua = a - b
    print(f"{a} - {b} = {ket_qua}")
elif ch == '*':
    ket_qua = a * b
    print(f"{a} * {b} = {ket_qua}")
elif ch == '/':
    if b != 0:  # Kiểm tra chia cho 0
        ket_qua = a / b
        print(f"{a} / {b} = {ket_qua}")
    else:
        print("Lỗi: Không thể chia cho 0!")
else:
    print(f"Ký tự '{ch}' không phải là một toán tử hợp lệ.")
```

```python
#cau10
import math

#Nhập x và n
x = float(input("Nhập x: "))
n = int(input("Nhập n: "))

#Tính S(x, n)
S = 0
for i in range(1, n + 1):
    S += x**i / math.factorial(i)

print("Giá trị S(x, n) =", S)
```

```python
#cau18
canh = int(input("Nhập độ dài cạnh hình vuông: "))
for i in range(canh):
    if i == 0 or i == canh - 1:
        print('* ' * canh)
    else:
        print('* ' + '  ' * (canh - 2 )+ '*')

chieu_cao = int(input("Nhập chiều cao tam giác: "))
for i in range(1, chieu_cao + 1):
    print('  ' * (chieu_cao - i) + '* ' * i)

print("\nHình tam giác vuông có 2 đỉnh chạm nhau:")
rows = 7  # số dòng

for i in range(rows):
    if i == 0:
        print("*")                    # dòng đầu có 1 *
    elif i == 1:
        print("* "," " * (i-1), "*", sep="")   # dòng 2 có 2 *
    elif i == 2:
        print("*  ", " " * (i-1), "*", sep="")   # dòng 3 có 2 *
    elif i == 3:
        print("* " * 7)                # dòng ngang dài
    elif i == 4:
        print(" " * (i+1), "  *   *")        # dòng 5 có 2 *
    elif i == 5:
        print(" " * (i+2), "  * *")         # dòng 6 có 2 *
    elif i == 6:
        print(" " * (i+3), "  *")          # dòng cuối có 1 *
```

```python
#cau19
import math

# Nhập dữ liệu
x = float(input("Nhập x (số thực): "))
n = int(input("Nhập n (số nguyên không âm): "))

S = 0
for k in range(n+1):   # k chạy từ 0 → n
    S += (x**(2*k+1)) / math.factorial(2*k+1)

print("Giá trị S(x,n) =", S)
```
